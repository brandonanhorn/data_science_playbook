```python
from tensorflow import keras
fashion_mnist = keras.datasets.fashion_mnist

(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
```

```python
from PIL import Image

for i, nppic in enumerate(X_train[:1000]):
    im = Image.fromarray(nppic)
    im.save(f"fashion/{i}-{y_train[i]}.png")
```

```python
from PIL import Image
import numpy as np
import re
import glob

X = []
y = []

files = glob.glob("fashion/*.png")

for file in files:
    img = np.asarray(Image.open(file))
    X.append(img)
    label = re.search("-(.*)\.", file).group(1)
    y.append(label)
    
X = np.array(X)
y = np.array(y)
```
