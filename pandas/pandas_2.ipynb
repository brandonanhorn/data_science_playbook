```python
import pandas as pd
```

```python
df = pd.read_csv("data/diamonds.csv")
```

# Basic analysis: show > 2.00 carats

```python
df[df.carat > 2.0]
```

# Count how many > 2.0 carats

```python
df[df.carat > 2.0].count()
```

# Sum how many have > 2.0 carats

```python
(df["carat"] > 2.0).sum()
```

# Show the first 3 > 2.0

```python
df[df.carat > 2.0][:3]
```

# Show the max > 2.0

```python
df[df.carat > 2.0].max()
```

# Show < 1.0 carat and a Premium cut diamond

```python
(df[(df.carat < 1.0) & (df.cut == "Premium")])
```
